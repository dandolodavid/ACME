```python
import pandas as pd
import numpy as np
from statwolfml import Builder
from ACME import ACME 
```

```python
glass_data = pd.read_csv('https://datahub.io/machine-learning/glass/r/glass.csv')
```

```python
glass_data['anomaly'] = 0
glass_data.loc[glass_data['Type'] == "headlamps",'anomaly']=1
glass_data.drop(columns='Type',inplace=True)
features = glass_data.drop(columns='anomaly').columns.to_list()
```

```python
model= 'logistic_regression'
build_model = Builder(model).feature_names(features).labels(['anomaly']).build()
train_model = build_model.train(glass_data)
```

```python
model = train_model.base_model()['model']
```

```python
acme_rf = ACME(model, 'anomaly', features = features, K=50, task = 'class')
```

```python
acme_rf = acme_rf.fit(glass_data,label_class=1)
```

```python
acme_rf.summary_plot()
```

```python
acme_local = acme_rf.fit_local(glass_data, local=1)
```

```python
acme_local.summary_plot()
```
