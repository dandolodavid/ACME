```python
from sklearn.inspection import plot_partial_dependence
import pandas as pd
import numpy as np
import time
import warnings
import shap
import matplotlib.pyplot as plt
```

```python
from sklearn import datasets
boston = datasets.load_boston()
X = boston.data
y = boston.target
        
dataframe = pd.DataFrame(X, columns=boston.feature_names)
dataframe['target'] = y
```

```python
features = dataframe.drop(columns={'target'}).columns
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor().fit(X,y)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize = (20,20))
fig = plot_partial_dependence(rfr, dataframe.drop(columns='target'), np.arange(0,len(features)), feature_names=features, fig = fig, grid_resolution=500) 
```
